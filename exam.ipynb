```python
women = train_df.loc[train_df.Sex == 'female']["Survived"] 
rate_women = sum(women)/len(women)

print("% of women who survived:", rate_women)
```

```python
men = train_df.loc[train_df.Sex == 'male']["Survived"]
rate_men = sum(men)/len(men)

print("% of men who survived:", rate_men)
```

```python
children = train_df.loc[train_df.Age < 19]["Survived"] 
rate_children = sum(children)/len(children)

print("% of children who survived:", rate_children)
```

```python
adult = train_df.loc[train_df.Age > 19]["Survived"] & train_df.loc[train_df.Age < 45]["Survived"]
rate_adult = sum(adult)/len(adult)

print("% of adult who survived:", rate_adult)
```

```python
old_people = train_df.loc[train_df.Age > 45]["Survived"] 
rate_old_people = sum(old_people)/len(old_people)

print("% of old_people who survived:", rate_old_people)
```
